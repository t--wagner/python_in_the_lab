```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import os

%config IPCompleter.greedy = True

%matplotlib inline
plt.style.use('seaborn-notebook')
```

```python
plt.style.use('seaborn-poster')
```

```python
# Load filenames
filenames = sorted(glob.glob('quantum_hall_data/*'))

fig, axs = plt.subplots(1, 2, figsize=(20, 8))
         
for filename in filenames:
    ary = np.genfromtxt(filename, skip_header=0)
    b, rxx, rxy = ary.T
        
    gate_voltage = filename.split('_')[-1][:-4]
        
    axs[0].plot(b, -2 * rxx, '-', label=gate_voltage)
    axs[1].plot(b, 2 * rxy, '-', label=gate_voltage)
    
for n in range(2, 7):
    axs[0].axhline(25812 / n, color='black', ls='--', lw=0.5)

axs[0].set_ylim(0, 25000)
axs[0].set_title('Hall resistance')
axs[0].set_xlabel('magnetic field B (T)')
axs[0].set_ylabel('Rxx (Ohm)')
axs[0].legend(loc=2)

axs[1].set_ylim(0, 7000)
axs[1].set_title('Longitudinal resistance') 
axs[1].set_xlabel('magnetic field B (T)')
axs[1].set_ylabel('Rxy (Ohm)')
axs[1].legend(loc=2);
```
